Consider the credit card dataset which contains the following columns:

**CLIENTNUM**: Primary key of the dataset

**Attrition_Flag**: Indicates if a customer is retained or attrited

**Customer_Age**: Age of the customer

**Gender**: Gender of the customer

**Dependent_count**: Number of people dependent on the customer

**Education_Level**: Highest level of education of the customer

**Income_Category**: Range of income of the customer

**Credit_Limit**: Credit card limit

**Total_Revolving_Bal**: Pending balance of the credit

**Avg_Purchase**: Amount of purchase made by the customer on credit card

**Total_Trans_Amt**: Total transaction amount

```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv('CreditCard_DV.csv')
df.head()
```

Based on the above information, import relevant libraries, and perform the following steps:

1. Create a bivariate plot to find if there is a correlation between credit card limit and average purchase made on the card.

2. Visualise the distribution of values for credit card limit and average purchase made on the card. Also, identify the outliers in the data, if any.

3. Provide a visual representation of the number of customers in each income group using a bar chart.

4. Plot the frequency distribution of the total transaction amount.

5. Graphically represent the percentage of customers retained and those attrited. Highlight the latter by slicing it apart from the main pie.

```python
# Creating a bivariate plot to find if there is a correlation between credit card limit and average purchase made on the card

ax = df.plot(["Credit_Limit"],["Avg_Purchase"],kind="scatter", color = "green",marker = "*",figsize=(7,5))
#To add labels and title to the output
ax.set_xlabel("Credit_Limit")	#sets label for x-axis
ax.set_ylabel("Avg_Purchase")	#sets label for y-axis
ax.set_title("Credit Card Limit vs Average Purchase",fontsize=16)	#sets title for the graph
plt.show()
```

There exists strong linear correlation between credit card limit and average puchase made on the card.

```python
# Visualising the distribution of values for credit card limit and average purchase made on the card.

plt.figure(figsize=(7, 3))
plt.subplot(1, 2, 1)
df["Credit_Limit"].plot(kind="hist",color="C2")
plt.title("Distribution of Credit_Limit", fontsize=16)
plt.legend()
plt.subplot(1, 2, 2)
df["Avg_Purchase"].plot(kind="hist",color="C2")
plt.title("Distribution of Avg_Purchase", fontsize=16)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Visualising for any outliers in the data using box plot

#The following lines enable us to use subplot functionality
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.set_figwidth(7)
fig.set_figheight(3)

#The following lines of code change the alignment from vertical to horizontal
bp = ax1.boxplot(df["Credit_Limit"],vert=False,patch_artist=True)
bp1 = ax2.boxplot(df["Avg_Purchase"],vert=False,patch_artist=True)

#The following lines of code are used to add labels to axes and title to the graph
ax1.set_title('Credit_Limit')
ax2.set_title('Avg_Purchase')

# To give color to box plot
colors = 'lightgreen'

for box in (bp['boxes'] + bp1['boxes']):
    box.set(facecolor=colors)

#In case of any superimposition of the subplots, the following functions caters the aesthetics
fig.tight_layout()
plt.show()
```

There are outliers in both Credit card limit and average purchases made in the card.

```python
# Visual representation of the number of customers in each income group using a bar chart

customer_counts = df['Income_Category'].value_counts()

plt.bar(customer_counts.index,customer_counts.values,color='green')
plt.xlabel('Income Category')
plt.ylabel('Number of Customers')
plt.title('Number of Customers in Each Income Category')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

```python
# Plot the frequency distribution of Total Transaction Amount
plt.figure(figsize=(5, 4))
df["Total_Trans_Amt"].plot(kind="hist",color="C2")
plt.title("Frequency Distribution of Total Transaction Amount", fontsize=15)
plt.legend()
plt.show()
```

```python
#Graphically representing the percentage of customers retained and those attrited


attrition_counts = df['Attrition_Flag'].value_counts() # To calculate count on each category

explode = (0, 0.2) #defining slice to emphasize , in this second slice

colors = ['#006400', '#66b3ff'] #to specifically give colors to slices

plt.pie(attrition_counts,labels= attrition_counts.index,autopct='%1.0f%%', startangle=20,explode=explode,colors=colors)

plt.show()
```

From pie chart we can summarise 93 % of customers has been retained and 7 % customers are attrited.
